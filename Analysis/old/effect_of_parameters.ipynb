# Analyzing the effects of parameters

first we need to load all the required modules

```python
import pickle
import matplotlib.pyplot as plt
```

reading the scores calculated

```python
def plot_data(axs, y_axis, x_axis, library):
    with open(library +  "_run_scores.pkl", "rb") as f:
        df = pickle.load(f)
    for i in range(2):
        for j in range(2):
            df[i][j].plot.scatter(ax = axs[i][j], x=x_axis, y=y_axis, title=library + " peak_presence_only:" + str(i == 0) + " combine:" + str(j==0))
    
    
```

The effect of mass:

```python
libraries = ["GNPS-MSMLS", "GNPS-NIH-NATURALPRODUCTSLIBRARY_ROUND2_POSITIVE", "GNPS-LIBRARY"]

# create 4 plots
fig, axs = plt.subplots(2, 2, figsize=(25, 15))
plot_data(axs, "score", "delta_mass", libraries[1])
```

the effect of # matched peaks:

```python
fig, axs = plt.subplots(2, 2, figsize=(25, 15))
plot_data(axs, "score", "#_matched_peaks", libraries[1])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(25, 15))
plot_data(axs, "score", "#_shifted_peaks", libraries[1])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(25, 15))
plot_data(axs, "score", "#_unshifted_peaks", libraries[1])
```

```python
import seaborn as sns
import pandas as pd

library = libraries[1]
with open(library +  "_run_scores.pkl", "rb") as f:
    df = pickle.load(f)
data = pd.DataFrame(columns=["library", "category", "value", "arg"])
categories = ['score']#, 'Is_Max']
for i in range(2):
    for j in range(2):
        arg = "peak_presence_only:" + str(i == 0) + " combine:" + str(j==0)
        for category in categories:
            data = data.append(pd.DataFrame({"library": libraries[0], "category": category, "value": df[i][j][category], "arg": arg}), ignore_index=True)
            print(arg, len(data))


sns.boxplot(x = data['category'], y = data['value'], hue = data['arg'])
```
